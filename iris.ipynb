```python
from sklearn.datasets import load_iris
import pandas as pd
```

```python
data = load_iris()
help(load_iris)
```

```python
data_arr = data.data
```

```python
f_n = data.feature_names
```

```python
data_df = pd.DataFrame()
cuenta = 0
for i in f_n:
    data_df[i] = data_arr[:,cuenta]
    cuenta+=1
print(data_df)
```

```python
data_df.plot.scatter(x = "sepal length (cm)", y = "sepal width (cm)")
data_df.plot.scatter(x = "sepal length (cm)", y = "petal length (cm)")
data_df.plot.scatter(x = "sepal length (cm)", y = "petal width (cm)")
data_df.plot.scatter(x = "sepal width (cm)", y = "petal length (cm)")
data_df.plot.scatter(x = "petal length (cm)", y = "petal width (cm)")
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(data_arr, test_size = 0.8)
```

```python
from sklearn.cluster import KMeans
```

```python
modelo = KMeans(n_clusters = 3, max_iter = 100, random_state = 42)
```

```python
modelo.fit(train)
```

```python
clusters = modelo.predict(test)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(test[:,2],test[:,3],c = clusters, cmap = "rainbow")
```
